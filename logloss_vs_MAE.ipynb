```python
from sklearn.metrics import log_loss, mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data
y_true = np.repeat([0,1], [3,7])
p = np.arange(0.01, 1, 0.01)

# Calculate logloss and MAE
logloss = []
mae = []
for i in p:
    y_pred = np.repeat(i, 10)
    logloss.append(log_loss(y_true, y_pred))
    mae.append(mean_absolute_error(y_true, y_pred))
    
# Plot the errors
plt.plot(p, logloss, label="logloss")
plt.plot(p, mae, label="mae")
plt.axvline(0.7, linestyle='--',color='r', label="0.7")
plt.legend()
```
